### Google BERT Confusion Matrix

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
model_names = ["Google BERT Base Uncased", "Google BERT Base Cased", "Google BERT Large Uncased", "Google BERT Large Cased"]

# Confusion matrix data
confusion_matrices = {
    "Google BERT Base Uncased": {"TP": 2309, "TN": 3822, "FP": 602, "FN": 470},
    "Google BERT Base Cased": {"TP": 2338, "TN": 3715, "FP": 709, "FN": 441},
    "Google BERT Large Uncased": {"TP": 2363, "TN": 4116, "FP": 308, "FN": 416},
    "Google BERT Large Cased": {"TP": 2368, "TN": 4122, "FP": 302, "FN": 411}
}

# Plotting the confusion matrices
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # Adjusted figure size

for i, model_name in enumerate(model_names):
    matrix = confusion_matrices[model_name]
    cm = np.array([[matrix["TP"], matrix["FN"]],
                   [matrix["FP"], matrix["TN"]]])
    
    ax = axs[i//2, i%2]
    cax = ax.matshow(cm, cmap=plt.cm.Blues, alpha=0.75)
    for (j, k), val in np.ndenumerate(cm):
        color = 'white' if (j == 0 and k == 0) or (j == 1 and k == 1) else 'black'
        ax.text(k, j, f"{val}", ha='center', va='center', color=color)
        
    ax.set_title(model_name, fontsize=10)  # Adjusted font size
    ax.set_xlabel("Predicted Label")
    ax.set_ylabel("True Label")
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_xticklabels(['Machine', 'Human'])
    ax.set_yticklabels(['Machine', 'Human'])

# Adjust the layout to make space for the central colorbar
fig.subplots_adjust(right=0.75, hspace=0.30, wspace=0.40)  # Adjusted spacing

# Add a single colorbar for all subplots
cbar_ax = fig.add_axes([0.78, 0.15, 0.03, 0.7])  # [left, bottom, width, height]
fig.colorbar(cax, cax=cbar_ax, orientation='vertical')

# Saving the plot as an image
plt.savefig("./images/Google_BERT_Confusion_Matrix.png", bbox_inches='tight')
plt.show()
```

### Meta Llama 2 Confusion Matrix

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
model_names = ["Meta Llama 2 7b", "Meta Llama 2 13b", "Meta Llama 2 70b"]

# Confusion matrix data
confusion_matrices = {
    "Meta Llama 2 7b": {"TP": 2763, "TN": 4416, "FP": 8, "FN": 16},
    "Meta Llama 2 13b": {"TP": 2768, "TN": 4414, "FP": 10, "FN": 11},
    "Meta Llama 2 70b": {"TP": 2778, "TN": 4423, "FP": 1, "FN": 1}
}

# Plotting the confusion matrices
fig, axs = plt.subplots(1, 3, figsize=(12, 4))  # Adjusted to 1 row and 3 columns

for i, model_name in enumerate(model_names):
    matrix = confusion_matrices[model_name]
    cm = np.array([[matrix["TP"], matrix["FN"]],
                   [matrix["FP"], matrix["TN"]]])
    
    ax = axs[i]
    cax = ax.matshow(cm, cmap=plt.cm.Blues, alpha=0.75)
    for (j, k), val in np.ndenumerate(cm):
        color = 'white' if (j == 0 and k == 0) or (j == 1 and k == 1) else 'black'
        ax.text(k, j, f"{val}", ha='center', va='center', color=color)
        
    ax.set_title(model_name, fontsize=12)  # Adjusted font size
    ax.set_xlabel("Predicted Label")
    ax.set_ylabel("True Label")
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_xticklabels(['Machine', 'Human'])
    ax.set_yticklabels(['Machine', 'Human'])

# Adjust the layout to make space for the central colorbar
fig.subplots_adjust(bottom=0.2, hspace=0.5, wspace=0.5)  # Adjusted spacing

# Add a single colorbar for all subplots
cbar_ax = fig.add_axes([0.15, 0.05, 0.7, 0.03])  # [left, bottom, width, height]
fig.colorbar(cax, cax=cbar_ax, orientation='horizontal')

# Saving the plot as an image
plt.savefig("./images/Meta_Llama2_Confusion_Matrix.png", bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Model names
model_names = ["Gemma 2b", "Gemma 7b"]
file_names = ["gemma-2b", "gemma-7b"]
colors = ["blue", "green"]

# Reading in the Checkpoint Data
dfs = []
for file_name in file_names:
    file_path = f"./images/{file_name}_Checkpoint_Data.csv"
    df = pd.read_csv(file_path)
    dfs.append(df)

# Plotting the graphs
plt.figure(figsize=(14, 8))

# Training Loss
plt.subplot(2, 2, 1)
for i, df in enumerate(dfs):
    plt.plot(df["Step"], df["Train Loss"], marker='o', color=colors[i], label=model_names[i], alpha=0.5, markersize=6)
plt.title("Training Loss")
plt.xlabel("Checkpoint Step")
plt.ylabel("Loss")
plt.legend()

# Evaluation Loss
plt.subplot(2, 2, 2)
for i, df in enumerate(dfs):
    plt.plot(df["Step"], df["Eval Loss"], marker='o', color=colors[i], label=model_names[i], alpha=0.5, markersize=6)
plt.title("Evaluation Loss")
plt.xlabel("Checkpoint Step")
plt.ylabel("Loss")
plt.legend()

# Accuracy
plt.subplot(2, 2, 3)
for i, df in enumerate(dfs):
    plt.plot(df["Step"], df["Accuracy"], marker='o', color=colors[i], label=model_names[i], alpha=0.5, markersize=6)
plt.title("Accuracy")
plt.xlabel("Checkpoint Step")
plt.ylabel("Accuracy")
plt.legend()

# F1-score
plt.subplot(2, 2, 4)
for i, df in enumerate(dfs):
    plt.plot(df["Step"], df["F1"], marker='o', color=colors[i], label=model_names[i], alpha=0.5, markersize=6)
plt.title("F1-score")
plt.xlabel("Checkpoint Step")
plt.ylabel("F1-score")
plt.legend()

# Adjust layout
plt.tight_layout()

# Saving the new plot as an image
plt.savefig("./images/Google_Gemma_Checkpoint_Plots.png", bbox_inches='tight')
plt.show()
```

### Google Gemma Confusion Matrix

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
model_names = ["Google Gemma 2b", "Google Gemma 7b"]

# Updated confusion matrix data
confusion_matrices = {
    "Google Gemma 2b": {"TP": 2735, "TN": 4393, "FP": 31, "FN": 44},
    "Google Gemma 7b": {"TP": 2766, "TN": 4422, "FP": 2, "FN": 13}
}

# Plotting the confusion matrices
fig, axs = plt.subplots(1, 2, figsize=(8, 4))  # Adjusted to 1 row and 2 columns

for i, model_name in enumerate(model_names):
    matrix = confusion_matrices[model_name]
    cm = np.array([[matrix["TP"], matrix["FN"]],
                   [matrix["FP"], matrix["TN"]]])
    
    ax = axs[i]
    cax = ax.matshow(cm, cmap=plt.cm.Blues, alpha=0.75)
    for (j, k), val in np.ndenumerate(cm):
        color = 'white' if (j == 0 and k == 0) or (j == 1 and k == 1) else 'black'
        ax.text(k, j, f"{val}", ha='center', va='center', color=color)
        
    ax.set_title(model_name, fontsize=12)  # Adjusted font size
    ax.set_xlabel("Predicted Label")
    ax.set_ylabel("True Label")
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_xticklabels(['Machine', 'Human'])
    ax.set_yticklabels(['Machine', 'Human'])

# Adjust the layout to make space for the central colorbar
fig.subplots_adjust(bottom=0.2, hspace=0.5, wspace=0.5)  # Adjusted spacing

# Add a single colorbar for all subplots
cbar_ax = fig.add_axes([0.15, 0.05, 0.7, 0.03])  # [left, bottom, width, height]
fig.colorbar(cax, cax=cbar_ax, orientation='horizontal')

# Saving the plot as an image
plt.savefig("./images/Google_Gemma_Confusion_Matrix.png", bbox_inches='tight')
plt.show()
```
